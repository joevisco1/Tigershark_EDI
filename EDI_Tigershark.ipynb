```python
import json
from tigershark.facade import f835, f837
from tigershark.parsers import M835_4010_X091_A1, M837_4010_X098_A1

def x12_to_json(x12_string, message_type):
    """
    Convert X12 string to JSON based on message type.

    Supported message types: '835', '837'
    """
    if message_type == '835':
        parsed = M835_4010_X091_A1().parse(x12_string)
        facade = f835(parsed)
    elif message_type == '837':
        parsed = M837_4010_X098_A1().parse(x12_string)
        facade = f837(parsed)
    else:
        raise ValueError(f"Unsupported message type: {message_type}")

    return json.dumps(facade.to_dict(), indent=2)




```
